1. Let $\textbf{u} = (1, -2, 4)$ and $\textbf{v} = (3,5,1)$ $\textbf{w} = (2,1,-3)$ find:<br>
a. $3\textbf{u}-2\textbf{v}$<br>
b. $5\textbf{u}+3\textbf{v}-4\textbf{w}$<br>
c. $\textbf{u} \cdotp \textbf{v}$,$\textbf{v} \cdotp \textbf{w}$, $\textbf{u} \cdotp \textbf{w}$<br>
d. ||$\textbf{u}||, ||$\textbf{v}||, ||$\textbf{w}$||<br>
e. If $\theta$ is the angle between $\textbf{u}$  and   $\textbf{v}$, find $\cos{\theta}$.<br>
f. If $\alpha$ is the angle between $\textbf{v}$  and   $\textbf{w}$, find $\cos{\alpha}$.<br>
g. If $\beta$ is the angle between $\textbf{u}$  and   $\textbf{w}$, find $\cos{\beta}$.<br>
h. Find $d(\textbf{u},\textbf{v})$.<br>
i. Find $d(\textbf{w},\textbf{v})$.<br>
j. Find $d(\textbf{u},\textbf{w})$. <br>
k. Find $proj(\textbf{u},\textbf{v})$.<br>
l. Find $proj(\textbf{w},\textbf{v})$.<br>
m. Find $proj(\textbf{u},\textbf{w})$. <br>  
<br>
<br>

```python
import numpy as np

def perform_operations():
    u = np.array([1, -2, 4])
    v = np.array([3, 5, 1])
    w = np.array([2, 1, -3])

    # Letter A
    result_a = (3*u) - (2*v)
    print("Letter A - 3ùêÆ-2v: ", result_a)

    # Letter B
    result_b = (5*u) + (3*v) - (4*w)
    print("Letter B - 5ùêÆ+3v-4w: ", result_b)

    # Letter C
    print("Letter C - u.v = ", u.dot(v))
    print("Letter C - v.w = ", v.dot(w))
    print("Letter C - u.w = ", u.dot(w))

    # Letter D
    print("Letter D - ||u||: ", np.sqrt(u.dot(u)))
    print("Letter D - ||v||: ", np.sqrt(v.dot(v)))
    print("Letter D - ||w||: ", np.sqrt(w.dot(w)))

    # Letter E
    angle_uv = u.dot(v) / (np.linalg.norm(u) * np.linalg.norm(v))
    print("Letter E - angle(u, v):", angle_uv)

    # Letter F
    angle_vw = v.dot(w) / (np.linalg.norm(v) * np.linalg.norm(w))
    print("Letter F - angle(v, w):", angle_vw)

    # Letter G
    angle_uw = u.dot(w) / (np.linalg.norm(u) * np.linalg.norm(w))
    print("Letter G - angle(u, w):", angle_uw)

    # Letter H
    print("Letter H - d(u, v): ", np.linalg.norm(u - v))

    # Letter I
    print("Letter I - d(w, v): ", np.linalg.norm(w - v))

    # Letter J
    print("Letter J - d(u, w): ", np.linalg.norm(u - w))

    # Letter K
    proj_uv = u.dot(v) * v / (np.linalg.norm(v) ** 2)
    print("Letter K - proj(u, v): ", proj_uv)

    # Letter L
    proj_wv = w.dot(v) * v / (np.linalg.norm(v) ** 2)
    print("Letter L - proj(w, v): ", proj_wv)

    # Letter M
    proj_uw = u.dot(w) * w / (np.linalg.norm(w) ** 2)
    print("Letter M - proj(u, w): ", proj_uw)

# Call the function
perform_operations()
```

5. Let $\textbf{u} = (1,2,-2)$,$\textbf{v} = (3,-12,4)$ and a scalar $k = -3$: <br>
Find $||u||, ||v||, ||u+v||, ||ku||$.<br>
<br>
<br>

```python
import numpy as np

def perform_operations():
    u = np.array([1, 2, -2])
    v = np.array([3, -12, 4])
    k = -3
    l = k * u
    w = u + v

    print('||u||: ', np.sqrt(u.dot(u)))
    print('||v||: ', np.sqrt(v.dot(v)))
    print('||u+v||: ', np.sqrt(w.dot(w)))
    print('||ku||: ', np.sqrt(l.dot(l)))

# Call the function
perform_operations()
```
