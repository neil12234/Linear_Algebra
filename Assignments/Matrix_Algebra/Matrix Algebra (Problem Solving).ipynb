1. Solve $X$ given that:
\begin{equation}
A =
\begin{bmatrix}

3 & 4 \\
\end{bmatrix},
B =
\begin{bmatrix}
-1 & 0 \\

\end{bmatrix}
\end{equation}
- $3X = A-2B$
- $2(A-B+2X) = 3(X-B)$
<br>
<br>

- Solve for $X$ in the equation $3X = A-2B$:

 \
     $2B$ = \begin{bmatrix} -2 & 0 \\ 2 & 2 \end{bmatrix}
     

\
     $A - 2B$ =
     \begin{bmatrix}
     3 & 2 \\
     1 & 2 \\
     \end{bmatrix}
     

 \
    $X$ =
     \begin{equation}
      \frac{1}{3} \times
     \begin {bmatrix}

     \begin{bmatrix}

     \frac{1}{3} & \frac{2}{3} \\
     \end{bmatrix}
     \end{equation}
     

 \
     $Final Answer$ =
     \begin{equation}
      X = 
     \begin{bmatrix}
     1 & \frac{2}{3} \\

     \end{bmatrix}

```python
# Python Solution
import numpy as np

# Define matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[-1, 0], [1, 1]])

# Compute A - 2B
result = A - 2 * B

# Divide result by 3
result = result / 3

# Print result
print(result)
```
